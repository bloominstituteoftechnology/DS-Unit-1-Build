```
import pandas as pd
play = pd.read_csv('player_playoffs_career.txt', engine='python')
print(play.shape)
play.head(20)
```

```
reg = pd.read_csv('player_regular_season_career.txt', engine='python')
print(reg.shape)
reg.head(20)
```

```
reg.dtypes
```

```
reg = reg.drop(reg[reg.leag == 'A'].index)
```

```
reg.head(20)
```

```
play = play.drop(play[play.leag == 'A'].index)
```

```
play.head(20)
```

```
play.rename(columns = {'gp':'playoffs gp', 'minutes':'playoff minutes', 'minutes':'playoff minutes', 'pts':'playoff pts', 'dreb':'playoff dreb',
                       'oreb':'playoff oreb', 'reb':'playoff reb', 'asts':'playoff asts', 'stl':'playoff stl', 'blk':'playoff blk', 
                       'turnover':'playoff turnover', 'pf':'playoff pf', 'fga':'playoff fga', 'fgm':'playoff fgm', 'fta':'playoff fta', 
                       'ftm':'playoff ftm',}, inplace = True)
```

```
play = play.drop(['tpa', 'tpm'], axis=1)
reg = reg.drop(['tpa', 'tpm'], axis=1)
play.head()
```

```
reg.head()
```

```
play.shape
```

```
reg.shape
```

```
reg['fg %'] = reg['fgm']/reg['fga']
reg['ft %'] = reg['ftm']/reg['fta']
play['playoff fg %'] = play['playoff fgm']/play['playoff fga']
play['playoff ft %'] = play['playoff ftm']/play['playoff fta']
```

```
play = play.drop(['playoff fgm', 'playoff fga'], axis=1)
reg = reg.drop(['fgm', 'fga'], axis=1)
```

```
play = play.drop(['playoff ftm', 'playoff fta'], axis=1)
reg = reg.drop(['ftm', 'fta'], axis=1)
```

```
play = play.drop(['leag'], axis=1)
reg = reg.drop(['leag'], axis=1)
```

```
play = play.reset_index(drop=True)
play.head(15)
```

```
reg = reg.reset_index(drop=True)
reg.head(9)
```

```
play.head(9)
```

```
q = pd.merge(play, reg, on=['firstname', 'lastname'])
q.head()
```

```
q.head(10)
```

```
q.shape
```

```
del q['ilkid_y']
```

```
q.rename(columns = {'ilkid_x':'ilkid'}, inplace = True)
```

```
q.head()
```

```
q['playoff pts/gp'] = q['playoff pts']/q['playoffs gp']
q['pts/gp'] = q['pts']/q['gp']
q['playoff reb/gp'] = q['playoff reb']/q['playoffs gp']
q['reb/gp'] = q['reb']/q['gp']
q['playoff asts/gp'] = q['playoff asts']/q['playoffs gp']
q['asts/gp'] = q['asts']/q['gp']
q['playoff stl/gp'] = q['playoff stl']/q['playoffs gp']
q['stl/gp'] = q['stl']/q['gp']
q['playoff blk/gp'] = q['playoff blk']/q['playoffs gp']
q['blk/gp'] = q['blk']/q['gp']
```

```
df = q[['ilkid',	'firstname', 'lastname',	'pts/gp',	'reb/gp',	'asts/gp',	'stl/gp',	'blk/gp',	'fg %',	'ft %', 'playoffs gp', 'playoff pts/gp',	'playoff reb/gp',	'playoff asts/gp',	'playoff stl/gp',	'playoff blk/gp',	'playoff fg %',	'playoff ft %']]
```

```
df.head()
```

```
df['fg %'] = df['fg %']*100
df['ft %'] = df['ft %']*100
df['playoff fg %'] = df['playoff fg %']*100
df['playoff ft %'] = df['playoff ft %']*100
```

```
df.head()
```

```
df.loc[df['firstname']=='Bernard']
```

```
import numpy as np
p = np.mean(df['pts/gp'])
pp = np.mean(df['playoff pts/gp'])
r = np.mean(df['reb/gp'])
pr = np.mean(df['playoff reb/gp'])
a = np.mean(df['asts/gp'])
pa = np.mean(df['playoff asts/gp'])
s = np.mean(df['stl/gp'])
ps = np.mean(df['playoff stl/gp'])
b = np.mean(df['blk/gp'])
pb = np.mean(df['playoff blk/gp'])
f = np.mean(df['fg %'])
pf = np.mean(df['playoff fg %'])
```

```
import matplotlib.pyplot as plt

b=np.array([[p,pp],[r,pr],[a,pa],[s,ps],[b,pb],[f/10,pf/10]])
w=pd.DataFrame(b, columns=['Regular Season','Playoffs'], index=['pts/gp', 'reb/gp', 'asts/gp', 'stl/gp', 'blk/gp', 'fg %'])

w.plot(kind='bar')
# Turn on the grid
plt.minorticks_on()
plt.title('Regular Season vs Playoffs')

plt.show()
```
